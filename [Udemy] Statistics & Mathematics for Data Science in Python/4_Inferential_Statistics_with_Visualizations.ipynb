```python
#common imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import scipy.stats as stats
%matplotlib inline

sns.set(color_codes = True)
```

```python
df_titanic = sns.load_dataset("titanic")
df_iris = sns.load_dataset("iris")
df_tips = sns.load_dataset("tips")
```
