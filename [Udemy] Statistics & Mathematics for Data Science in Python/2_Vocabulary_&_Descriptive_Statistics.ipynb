```python

df_wesr[['W', 'L', 'W/L%', 'GB', 'PS/G', 'PA/G', 'SRS']] = df_wesr[['W', 'L', 'W/L%', 'GB', 'PS/G', 'PA/G', 'SRS']].apply(pd.to_numeric, errors="coerce")
```

```python
df_wesr.dtypes
```

```python
df_west = df_wesr.copy()
```

```python
df_west["W"].mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_west1 = df_west.select_dtypes(include=["number"])
cor = df_west1.corr()
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(cor, annot=True, linewidth=3, annot_kws={'size':15},
            cbar=False, cmap="coolwarm")

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)


# # Візуалізуємо кореляцію
# plt.figure(figsize=(10, 6))  # Розмір графіка
# sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# plt.title("Кореляційна матриця")
# plt.show()
```
