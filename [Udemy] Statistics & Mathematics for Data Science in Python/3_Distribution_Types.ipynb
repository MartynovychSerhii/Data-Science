```python
#working with binomial Distribution
#binomial(n = {number of trials}, p = {probability of success}, size = N {number of scenarios executed})
```

```python
binomial_data = random.binomial(n=10, p=0.99, size=20)
```

```python
binomial_data
```

```python
sns.distplot(random.binomial(n=100, p=0.5, size=1000), hist=True, kde=True)
```

```python
sns.distplot(random.binomial(n=100, p=0.2, size=1000), hist=True, kde=True)
```

```python
#working with the Poisson distribution
```

```python
#poisson(lambda = {n}, size = {N})
poisson_data = random.poisson(lam= 5, size=10)
```

```python
poisson_data
```

```python
sns.distplot(poisson_data, hist=True)
```

```python
sns.distplot(random.poisson(lam= 20, size=100), hist=True)
```

```python
url = 'https://www.basketball-reference.com/leagues/NBA_2025.html'
```

```python
df = pd.read_html(url, header=0)
```

```python
temp_df = df[3]
df_west = temp_df[~temp_df["W"].str.contains("Division")]
df_west1 = df_west.copy()
df_west1[['W', 'L', 'W/L%', 'GB', 'PS/G', 'PA/G', 'SRS']] = df_west1[['W', 'L', 'W/L%', 'GB', 'PS/G', 'PA/G', 'SRS']].apply(pd.to_numeric, errors="coerce")
```
